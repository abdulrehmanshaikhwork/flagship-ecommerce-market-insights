```python
# Importing
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-IN,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Connection": "keep-alive",
}


# Creating Empty List Named Items
items = []

# Start From Page 1
page = 1

# Runnig While Loop
while True:
    print(f"Page Scraping {page}")
    url = f"https://www.flipkart.com/search?q=smartphones&page={page}"

    # Creating Response And Soup
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.text, "html.parser")

    # Testing
    print(response.status_code)
    print(soup.title)


    # Finding Products
    products = soup.find_all("div", attrs={"data-id": True})

    # Execute If Products Not Found
    if not products:
        print("No More Product Found. Stoping.")
        break

    # Running For loop in Products
    for product in products:
        # Title
        product_title = product.find("div", class_="RG5Slk")
        title = product_title.text if product_title else None

        # Price
        product_price = product.find("div", class_="hZ3P6w DeU9vF")
        price = product_price.text if product_price else None

        # Discount
        discount_div = product.find("div", class_="HQe8jr")
        if discount_div:
            span = discount_div.find("span")
            discount = span.text.split()[0] if span else None
        else:
            discount = None

        # Rating Int
        rating_int = product.find("div", class_="MKiFS6")
        if rating_int:
            rating = rating_int.text
        else:
            rating = None

        # Rating Element
        rating_element = product.find("span", string=lambda x: x and "Ratings" in x)
        if rating_element:
            no_of_rating = rating_element.text.split()[0]
        else:
            no_of_rating = None

        # Reviews Element
        reviews_element = product.find("span", string=lambda x: x and "Reviews" in x)
        if reviews_element:
            no_of_reviews = reviews_element.text.split()[0]
        else:
            no_of_reviews = None

        # Appending All The Things In Empty List Of Items
        items.append([title,
                     price,
                     discount,
                     rating,
                     no_of_rating,
                     no_of_reviews])
        
    # Will Count Total Page Scrap So Far
    print(f"Total Page Scrap So Far {len(items)}")

    # Page Increment
    page += 1

    # For Anti-blocking
    time.sleep(1)
```
